```python
start_position = (0,0)
for i in range(lab_map.shape[0]):
    for j in range(lab_map.shape[1]):
        if lab_map[i,j] == '^':
            start_position = (i,j)
```

```python
start_position
```

```python
def check_out_of_map(next_position, lab_map):
    max_row, max_column = lab_map.shape[0], lab_map.shape[1]
    row, column = next_position
    if row == -1 or row == max_row or column == -1 or column == max_column:
        return True
    else:
        return False
```

```python
def moving(current_position, current_direction, lab_map):
    visited_map = np.zeros(lab_map.shape).astype(int)
    visited_map[current_position] = 1
    next_position = current_position
    counter = 1
    while check_out_of_map(next_position, lab_map) == False:
        current_row, current_col = current_position
        if current_direction == 'up':
            next_position = (current_row-1, current_col)
        elif current_direction == 'right':
            next_position = (current_row, current_col+1)
        elif current_direction == 'left':
            next_position = (current_row, current_col-1)
        elif current_direction == 'down':
            next_position = (current_row+1, current_col)
        
        if check_out_of_map(next_position, lab_map) == False:
            if  lab_map[next_position] == '.' or lab_map[next_position] == '^':
                current_position = next_position
                if visited_map[current_position] == 0:
                    visited_map[current_position] = 1
                    counter = counter + 1
            elif lab_map[next_position] == '#':
                if current_direction == 'up':
                    current_direction = 'right'
                elif current_direction ==  'right':
                    current_direction = 'down'
                elif current_direction ==  'down':
                    current_direction = 'left'
                elif current_direction ==  'left':
                    current_direction = 'up'
    return counter
```

```python
result = moving(start_position, 'up', lab_map)
print('Answer:', result)
```

```python
#check if current map having a loop
def check_have_a_loop(current_position, current_direction, lab_map):
    visited_map = [[{'up': False, 'down': False, 'left': False, 'right': False} \
                    for i in range(lab_map.shape[0])] for j in range(lab_map.shape[1])]
    visited_map[current_position[0]][current_position[1]]['up'] = True
    next_position = current_position
    visiting_status= [False]
    stuck_in_a_loop = False
    while check_out_of_map(next_position, lab_map) == False:
        current_row, current_col = current_position
        if current_direction == 'up':
            next_position = (current_row-1, current_col)
        elif current_direction == 'right':
            next_position = (current_row, current_col+1)
        elif current_direction == 'left':
            next_position = (current_row, current_col-1)
        elif current_direction == 'down':
            next_position = (current_row+1, current_col)
        
        if check_out_of_map(next_position, lab_map) == False:
            if  lab_map[next_position] == '.' or lab_map[next_position] == '^':
                current_position = next_position
                current_row, current_col = current_position
                if visited_map[current_row][current_col][current_direction] == False:
                    visited_map[current_row][current_col][current_direction] = True
                    visiting_status.append(False)
                else:
                    visiting_status.append(True)
                    
            elif lab_map[next_position] =='#' or lab_map[next_position] == 'O':
                if len(visiting_status) > 0 and np.array(visiting_status).all():
#                 if np.array(visiting_status).all():
                    stuck_in_a_loop = True
                    return stuck_in_a_loop
                if current_direction == 'up':
                    current_direction = 'right'
                elif current_direction ==  'right':
                    current_direction = 'down'
                elif current_direction ==  'down':
                    current_direction = 'left'
                elif current_direction ==  'left':
                    current_direction = 'up'
                visiting_status = []
    return stuck_in_a_loop
```
